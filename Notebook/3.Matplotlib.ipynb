```python
# Sử dụng matplotlib thì cần import
import matplotlib.pyplot as plt
import numpy as np
```

#### Vẽ đồ thị hàm bậc 2
$y=ax^2+bx+c$

```python
# Định nghĩa hàm bậc 2
def HamBacHai(a, b, c, x):
    return a * x ** 2 + b * x + c

# Gán giá trị cho các biến
a, b, c = 2, -3, 1

# Tạo điểm trên hoành độ
xpoints = np.arange(-10,10, step=0.1)

# Tính tung độ

ypoints = np.array([HamBacHai(a,b,c,x) for x in xpoints])
```

```python
plt.plot(xpoints, ypoints, label = "Hàm bậc 2", color = "red")
plt.grid() # Vẽ đường
plt.xlabel('x', loc='right')
plt.ylabel('y', loc='top')
plt.show()
```

#### Figures và Subplots
Tạo biểu đồ dưới dạng hướng đối tượng để dễ dàng xử lý hơn

Trong Matplotlib cái đối tượng hình ảnh thì gọi là Figures. Có thể điều chỉnh việc vẽ đối tượng một cách chính xác và hỗ trợ các nhu cầu phức tạp hơn so với dùng mặc định.

```python
# Tạo đối tượng figure
fig = plt.figure()
# Thêm hình vẽ con có 2 hàng và 2 cột, dùng ô thứ 1,2,3
ax1 = fig.add_subplot(2,2,1)
ax1.plot(np.random.randn(50).cumsum(), 'g--')
ax1.plot(np.random.randn(50).cumsum(), 'r--')
ax1.plot(np.random.randn(50).cumsum(), 'b--')
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)
plt.show()
```

#### Colors, Markers, Line Styles, Ticks, Labels and Lengends

```python
# g = green
# o = biểu tượng hiển thị ở điểm DL
# -- = vẽ nét đứt
plt.plot(np.random.randn(20).cumsum(),"r*--")
plt.grid() # Vẽ đường
plt.xlabel('x') # Thiết lập nhãn cho trục X
plt.ylabel('y') # Thiết lập nhãn cho trục Y
plt.title('Demo')   # Tiêu đề của đồ thị
plt.xticks(np.arange(0,21,5))  # Thiết lập các giá trị cho trục X 
plt.show()
```
