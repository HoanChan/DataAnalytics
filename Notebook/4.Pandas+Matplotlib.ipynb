```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv('Data/diemthi_ltt_2022.csv')
df.head()
```

```python
# Lấy điểm số các môn
mon1 = df['Mon1']
mon2 = df['Mon2']
mon3 = df['Mon3']
```

#### Vẽ Histogram cho dữ liệu
Histogram dùng để hiển thị sự phân bố (Tần suất) của dữ liệu

```python
# Tạo mức điểm để đếm giá trị
bins = np.arange(0, 10.25, step = 0.25)
fig, axs = plt.subplots(3, 1, sharey = True, tight_layout = True, figsize = (10,10))
#  Lệnh hist sẽ đếm các giá trị trong bins từ bộ dữ liệu trong mon và vẽ đồ thị cột
axs[0].hist(mon1, bins=bins, alpha=0.8, color='b')
axs[0].title.set_text('Phân bố điểm môn 1')
axs[1].hist(mon2, bins=bins, alpha=0.8, color='g')
axs[1].title.set_text('Phân bố điểm môn 2')
axs[2].hist(mon3, bins=bins, alpha=0.8, color='r')
axs[2].title.set_text('Phân bố điểm môn 3')
```

```python
# Vẽ đường ước lượng mật độ
mon1.plot.density(color = 'r')
```

#### Scatter plot
Dùng để xem phân bố của các điểm dữ liệu, vẽ dựa vào toạ độ x và y

```python
# Load Iris datashet
df=pd.read_csv('data/iris.csv')
df.head()
```

```python
# Xem các thống kê cơ bản của tập dữ liệu
df.describe()
# mean = trung bình, std = độ lệch chuẩn
```

```python
# Lọc lấy dữ liệu
setosa = df[df.variety=='Setosa']
virginica = df[df.variety=='Virginica']
versicolor = df[df.variety=='Versicolor']
print(len(versicolor), len(virginica), len(setosa))
```

```python
fig, ax=plt.subplots()
fig.set_size_inches(13,7)
# Vẽ giá trị width và length của đặc trưng petal
ax.scatter(setosa['petal.length'], setosa['petal.width'], label="setosa", facecolor='blue')
ax.scatter(virginica['petal.length'], virginica['petal.width'], label="setosa", facecolor='red')
ax.scatter(versicolor['petal.length'], versicolor['petal.width'], label="setosa", facecolor='green')
```

```python
fig, ax=plt.subplots()
fig.set_size_inches(13,7)
# Vẽ giá trị width và length của đặc trưng sepal
ax.scatter(setosa['sepal.length'], setosa['sepal.width'], label="setosa", facecolor='blue')
ax.scatter(virginica['sepal.length'], virginica['sepal.width'], label="setosa", facecolor='red')
ax.scatter(versicolor['sepal.length'], versicolor['sepal.width'], label="setosa", facecolor='green')
```
