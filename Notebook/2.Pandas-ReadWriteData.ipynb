### Bài tập 2
Cho file `Paramter.csv` mỗi dòng chứa bộ hệ số của một phương trình $a^2 + bx + c = 0$

```python
import numpy as np
import pandas as pd

def Calc(a,b,c):
    if a == 0:
        if b == 0:
            if c == 0: return "Nghiệm tuỳ ý"
            return "Vô nghiệm"
        return "x = {}".format(-c/b)
    delta = b**2-4*a*c
    if delta < 0 : return 'Vô nghiệm'
    if delta == 0: return 'Nghiệm kép x = {}'.format(-b/2/a)
    x1 = (-b+np.sqrt(delta))/2/a
    x2 = (-b-np.sqrt(delta))/2/a
    return 'x1 = {}, x2 = {}'.format(x1,x2)

df=pd.read_csv('Data/parameters.csv')
# Đọc từng dòng giá trị và lấy từng cột giá trị
for idx, row in df.iterrows():
    print('Phương trình {}x^2 + {}x + {} = 0'.format(row['a'],row['b'],row['c']))
    print(Calc(row['a'],row['b'],row['c']))
    print('---------------------------------------')
```
